```python
import pandas as pd
url ='Salary_Data.csv'
data = pd.read_csv(url) # doc du lieu
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
data.hist()
plt.show
# ve do thij du lieu
```

```python
correlation = data.corr()
print(correlation)
# bản biểu 

import seaborn as sns
sns.pairplot(data)
sns.heatmap(correlation,annot=True)
data.plot(kind = 'scatter', x = 'Salary', y = 'YearsExperience', figsize = (9,9), color = 'black')
```

```python
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['Salary']), y =data['YearsExperience'])
print(model.intercept_)
print(model.coef_)
model.score(X = pd.DataFrame(data['Salary']), y =data['YearsExperience'])


```

```python
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['Salary']), y =data['YearsExperience'])
print(model.intercept_)
print(model.coef_)
model.score(X = pd.DataFrame(data['Salary']), y =data['YearsExperience'])
model = linear_model.LinearRegression()
X = pd.DataFrame([data['Salary'],data['Salary']**2]).T# chir thoong soos 
model.fit(X,y=data['YearsExperience'])
range_values = np.arange(1, 10, 1)
predictor = pd.DataFrame([range_values, range_values**2]).T
y_pred = model.predict( X= predictor)
plt.plot(range_values, y_pred, color='blue')
plt.show()

```
