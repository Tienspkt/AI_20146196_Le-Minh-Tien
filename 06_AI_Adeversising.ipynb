```python
import pandas as pd
url ='Advertising.csv'
data = pd.read_csv(url) # doc du lieu
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
data.hist()
plt.show
```

```python
correlation = data.corr()
print(correlation)
# bản biểu 

import seaborn as sns
sns.pairplot(data)
sns.heatmap(correlation,annot=True)
data.plot(kind = 'scatter', x = 'radio', y = 'sales', figsize = (9,9), color = 'black')
```

```python
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['radio']), y =data['sales'])
print(model.intercept_)
print(model.coef_)
model.score(X = pd.DataFrame(data['radio']), y =data['sales'])
model = linear_model.LinearRegression()
X = pd.DataFrame([data['radio'],data['radio']**2]).T# chir thoong soos 
model.fit(X,y=data['sales'])
print('Intercept cua mo  hinh : ',model.intercept_)
print('Coeffient cuar mo hinh : ',model.coef_)
print('Do chinh xac cua mo hinh : ',model.score(X,y = data['sales']))
range_values = np.arange(0, 25, 5)
data.plot(kind='scatter', x='radio', y='sales', figsize=(9, 9), color='black', xlim=(0, 7))
predictor = pd.DataFrame([range_values, range_values**2]).T # chinh thoong so theo baac
y_pred = model.predict(X=predictor)
plt.plot(range_values, y_pred, color='blue')
plt.show()
```
